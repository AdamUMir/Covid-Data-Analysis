```python
#Importing Libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Forming Dataframe
df = pd.read_csv('covid_vaccine_statewise.csv',header=[0]) #Reading Dataframe with header as first line
#Describing the Dataset
print("\nDataframe Shape:\n",df.shape) #Displaying Shape of Dataframe
print("\nDataframe Columns:\n",df.columns) #Displaying the Column Names
print("\nDataframe Datatypes:\n",df.dtypes) #Displaying the Coloumn Datatypes
print("\nDataframe Head:\n",df.head()) #Displaying first few "head" rows
print("\mDataframe Description Statistics:\n",df.describe(include='all'))
```

```python
df = df.loc[df['Updated On'] == '09/08/2021'] #Selecting Latest Updated On Rows to perform analytics on
print("\nChecking for Null Values if any:\n",df.isnull().sum()) #Checking for Null Values
print("\nDataframe shape:\n",df.shape)
```

```python
#Handling Null Values by replacing Vaccinated Columns with values of corresponding Doses Administered Columns.
df['18-44 Years(Individuals Vaccinated)'] = df['18-44 Years (Doses Administered)']
df['45-60 Years(Individuals Vaccinated)'] = df['45-60 Years (Doses Administered)']
df['60+ Years(Individuals Vaccinated)'] = df['60+ Years (Doses Administered)']
df['Male(Individuals Vaccinated)'] = df['Male (Doses Administered)']
df['Female(Individuals Vaccinated)'] = df['Female (Doses Administered)']
df['Transgender(Individuals Vaccinated)'] = df['Transgender (Doses Administered)']
df['Total Individuals Vaccinated'] = df['Male(Individuals Vaccinated)'] + df['Female(Individuals Vaccinated)'] + df['Transgender(Individuals Vaccinated)']
df.isnull().sum()
#Null Values have been handled
```

```python
#Number of stateswise persons adminsitered for first dose
print("Bar plot for First Dose stateswise:",sns.barplot(x = 'State',y = 'First Dose Administered', data = df[1:]))
```

```python
#Number of stateswise persons adminsitered for second dose
print("Bar plot for Second Dose stateswise:",sns.barplot(x = 'State',y = 'Second Dose Administered',data = df[1:]))
```

```python
#Dataframe for just India
dv=df.loc[df['State'] == 'India']
#Printing Number of Males and Females Vaccinated
print("Number of Males Vaccinated:\n",dv['Male(Individuals Vaccinated)'])
print("Number of Females Vaccinated:\n",dv['Female(Individuals Vaccinated)'])
```
