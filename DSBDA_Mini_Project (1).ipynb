```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AdamUMir/Covid-Data-Analysis/main/covid_vaccine_statewise.csv',header=[0]) #Reading Dataframe with header as first line
df.head()
```

```python
df.describe()
```

###TASK 2: NUMBER OF PERSONS STATEWISE VACCINATED FOR FIRST DOSE IN INDIA

```python
First_dose_statewise=df.groupby('State')[['First Dose Administered']].sum()
First_dose_statewise
```

```python
First_dose_statewise.drop(['India'], inplace=True)
plt.figure(figsize=(15,15))
plt.title("Number of Person Statewise Vaccinated for First Dose in India")
sns.barplot(x=First_dose_statewise.index, y=First_dose_statewise['First Dose Administered'])
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.grid("True")
plt.subplots_adjust(bottom=0.15)
plt.ylabel("Number of DOSES(MILLION)")
```

### TASK 3: NUMBER OF PERSON STATEWISE VACCINATED FOR SECOND DOSE IN INDIA

```python
Second_dose_statewise=df.groupby('State')[['Second Dose Administered']].sum()
Second_dose_statewise
```

```python
Second_dose_statewise.drop(['India'], inplace=True)
plt.figure(figsize=(15,15))
plt.title("Number of Person Statewise Vaccinated for Second Dose in India")
sns.barplot(x=Second_dose_statewise.index, y=Second_dose_statewise['Second Dose Administered'])
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.grid()
plt.subplots_adjust(bottom=0.15)
plt.ylabel("Number of DOSES(MILLION)")
```

###TASK 4: NUMBER OF MALES VACCINATED


```python
Males_vaccinated=df['Male (Doses Administered)'].sum()
Males_vaccinated
```

###TASK 5: NUMBER OF FEMALES VACCINATED

```python
Females_vaccinated = df['Female (Doses Administered)'].sum()
Females_vaccinated
```

```python
xlabels=["Males", "Females"]
ylabels=[Males_vaccinated, Females_vaccinated]
fig=plt.figure(figsize=(8,6))
plt.bar(xlabels,ylabels, color=["Blue", "Pink"])
plt.ylabel("NUMBER OF PEOPLE VACCINATED (MILLIONS)")
plt.title("NUMBER OF MALES AND FEMALES VACCINATED")
plt.show()
```
